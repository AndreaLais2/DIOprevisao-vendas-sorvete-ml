```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
dados = {
"temperatura": [18, 20, 22, 24, 26, 28, 30, 32, 34, 36],
"vendas": [ 120, 150, 180, 220, 260, 300, 340, 380, 42, 460]
}

df = pd.DataFrame(dados)

df
```

```python
plt.scatter(df["temperatura"], df["vendas"])
plt.xlabel("Temperatura")
plt.ylabel("Vendas de Sorvete")
plt.title("Relação entre Temperatura e Vendas")
plt.show()
```

```python
x = df[["temperatura"]]
y = df["vendas"]
```

```python
x_treino, x_teste, y_treino, y_teste = train_test_split(
x, y, test_size=0.2, random_state=42
)
```

```python
modelo = LinearRegression()

modelo.fit(x_treino, y_treino)
```

```python
previsao = modelo.predict([[35]])

print("Previsão de vendas:", round(previsao[0]))
```

```python
plt.scatter(x, y)
plt.plot(x, modelo.predict(x), color="red")
plt.xlabel("Temperatura")
plt.ylabel("Vendas")
plt.title("Modelo de Regressão Linear")
plt.show()
```
